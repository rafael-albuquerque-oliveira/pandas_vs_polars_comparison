```python
df_data_jan = pd.read_parquet('data/yellow_tripdata_2025-01.parquet', engine='pyarrow')
df_data_jul = pd.read_parquet('data/yellow_tripdata_2025-07.parquet', engine='pyarrow')
df_data_sep = pd.read_parquet('data/yellow_tripdata_2025-09.parquet', engine='pyarrow')

df_concat = pd.concat([df_data_jan, df_data_jul, df_data_sep])
```

```python
df_concat.to_csv('data/test_data.csv')
```

```python
path = 'data/test_data.csv'
df = pl.scan_csv(path)
df = df.filter(pl.col("passenger_count") > 1)
grouped_df = df.group_by("PULocationID").agg(pl.col("fare_amount").mean()).sort(by='fare_amount', descending=True).collect()
```
